# Downloading the data from static API from a styled map

```python
__author__ = 'SVM'
import warnings
warnings.filterwarnings("ignore")
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
import requests
# %matplotlib inline

def draw_patches_mapbox(city,patchid,lat,lng,whichmap='roads',sz=700,zoom=15,foldername='./img/'):
# whichmap = 'roadsnbuildings'
    if whichmap == 'roads':
        roadsonly = 'cix65j03d008v2pnw5npbx7qj'
        style_ID ='sevamoo/' + roadsonly
    if whichmap == 'buildings':
        buildings_only ='cix65dm69007v2pnol1tezogn'
        style_ID ='sevamoo/' + buildings_only
    if whichmap == 'roadsnbuildings':
        road_building = 'cix64z4rs008i2qnv5fv2sqmn'
        style_ID ='sevamoo/' + road_building

    if whichmap == 'Satellitenroads':
        Satellitenroads = 'cix8ruejj00c32pnov0dk2aeo'
        style_ID ='sevamoo/' + Satellitenroads

    lat= str(lat)
    lng = str(lng)
    zoom = str(zoom)
    imgsz =str(sz)+'x'+str(sz)+'@2x'
    token = 'pk.eyJ1Ijoic2V2YW1vbyIsImEiOiJXcTg2dEI4In0.t2FKdE5vgT-BihNJusEz6Q'

    url = 'https://api.mapbox.com/styles/v1/'+style_ID+'/static/'+lng+','+lat+','+zoom+'/'+imgsz +'?attribution=false&logo=false&'+'access_token='+ token

    res = requests.get(url,timeout=120)

    with open(foldername+whichmap+'/'+city+'_'+str(patchid)+'_'+str(lat).replace('.','-')+'_'+str(lng).replace('.','-')+'.png', 'wb') as output:
        _ = output.write(res.content)
    del res


# Numbeo cities with geocoding
# all_cities_centers = pd.read_csv('./data/numbeo_cities_center_bbox.csv')


# all cities with >75k cities
all_cities_centers = pd.read_csv('./data/all_cities_center_bbox.csv')

print all_cities_centers.shape
all_cities_centers.head()

import glob
import os
path = './all_cities/img_centers/roads/'
all_crawled_cities = []
counter = 0
for filename in glob.glob(os.path.join(path, '*.png')):
    
    filename = filename.replace(path,'').replace('_','').replace('-','').replace('.png','').replace("'",'')
    for dig in range(10):
        filename = filename.replace(str(dig),'')
    filename = filename.lower()
    all_crawled_cities.append(filename)

print len(all_crawled_cities)


# for these cities if they have not crawled previously in the other numbeo set, we just get the centers. Later to be used 
# for unsupervised learning tasks
for r in range(all_cities_centers.shape[0])[:]:
    
    imgsz = 1200
    zoom = 12.5
    howmanypatch = 100
    
    
    
    df = all_cities_centers.ix[r].copy()
    city_name = df.city
    country_name = df.country
    lat = df.lat
    lng = df.lng
    city = str(city_name) + " " + str(country_name)
    
    check = city
    if r%100 == 0:
        print r
    for dig in range(10):
        check = check.replace(str(dig),'')
    check = check.replace('_','').replace('-','').replace("'",'').lower()
    if check in all_crawled_cities:
        continue
    print r, city
    try:
        patchid = 0
        whichmap='roads'
        path = './all_cities/img_centers/'
        draw_patches_mapbox(city,patchid,lat,lng,whichmap=whichmap,sz=imgsz,zoom=zoom,foldername=path)
        
        #Now for all save roadsnbuildings
        whichmap='roadsnbuildings'
        zoom = 15.1
        imgsz = 1200
        path = './all_cities/img_centers/'
        draw_patches_mapbox(city,patchid,lat,lng,whichmap=whichmap,sz=imgsz,zoom=zoom,foldername=path)
    except:
        pass
```
