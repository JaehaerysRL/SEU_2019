# Data collection pipeline?

# 1- get the *.osm.pbf files for the region you want
* OSM extracts from here: https://download.geofabrik.de/
* Take the *.osm.pbf file

# 2- Filter the layers that you want (e.g. commercial buildings)
#### osmconvert
* convert .pbf or bz by https://wiki.openstreetmap.org/wiki/Osmconvert download the source
    * download it like this: wget -O - http://m.m.i24.cc/osmconvert.c | cc -x c - -lz -O3 -o osmconvert
* ./osmconvert netherlands-latest.osm.pbf >netherlands-latest.osm

### osmfilter 
* good guide: https://journocode.com/2018/01/08/extract-geodata-openstreetmap-osmfilter/
* ./osmfilter netherlands-latest.osm  --out-count=building |head

* ./osmfilter netherlands-latest.osm --keep="highway=primary =secondary waterway=river" >streets.osm
* ./osmfilter netherlands-latest.osm --keep=building=* -o=netherlands_building.osm
* ./osmfilter netherlands-latest.osm --keep="highway=primary =secondary waterway=river" >streets.osm

* ./osmfilter netherlands-latest.osm --keep="building=commercial =industrial =garages = warehouse =service =office =retail " -o=netherlands_building_commercial.osm

# 3- Convert  *.osm to *.geojson
* GDAL is installed in ogr conda environment
* ogr2ogr -f GeoJSON netherlands_building.geojson  netherlands_building.osm  (Dosnâ€™t work)
* ogr2ogr -f GeoJSON netherlands_building.geojson netherlands_building.osm multipolygons
* ogr2ogr -f GeoJSON netherlands_building_commercial.geojson netherlands_building_commercial.osm  multipolygons
* Tips: https://gis.stackexchange.com/questions/22788/how-do-you-convert-osm-xml-to-geojson/69622#69622


# 4- Selecting POIs from *.osm.pbf to *.csv
* POIs OsmPoisPbf 
* java -Xmx4g -jar osmpois.jar --relations --allowUnclosedWays --outputFile switzerland-latest-pois.csv switzerland-latest.osm.pbf

```python
# Now we qualify each of the existing POIs with their neighboring activities
```

```python
chosen_r = 500
howmany_keys = 50
Type_of_Search_for_POI = 'closest distance'
# Type_of_Search_for_POI = 'radius search'
keys = [str(p) for p in poi_groups.index.values[:howmany_keys]]
```

```python
import sklearn.neighbors as neighbors

# Let's focus on the most frequent POIs
# buildings = pd.read_csv(buildings_path)



pois_with_neigh = pois.copy()
for count, key in enumerate(keys):
   
    inds = list(np.where(pois['type'].values[:]==key)[0])
    print (count,key, len(inds))
    sel_poi = pois.loc[inds]
    coords_poi = sel_poi[['lat','lon']].values[:]
    
    
    if Type_of_Search_for_POI == 'radius search':
    
        #Link to Sklearn: http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.BallTree.html
        tree = neighbors.BallTree(np.radians(coords_poi), metric='haversine')     
        # Here we look on a spatial radius around each buildings
        # And simply count the number of each POI in that distance
        ms_per_radian = 6371008.8
#         r = 500 #in meters
        r_query = chosen_r/ms_per_radian
#         heat_buildings = tree.query_radius(np.radians(coords_buildings), r=r_query, count_only=True)
#         buildings[key] = heat_buildings
        pois_with_neigh[key] = tree.query_radius(np.radians(pois_with_neigh[['lat','lon']].values[:]), r=r_query, count_only=True)

    if Type_of_Search_for_POI == 'closest distance':
        # We could also ask what is the distance to the closest POI of type i
        # There, we need to store the data
        tree = neighbors.BallTree(np.radians(coords_poi), metric='haversine')
        dist, ind = tree.query(np.radians(pois_with_neigh[['lat','lon']].values[:]),k=1)
        ms_per_radian = 6371008.8 #converting the radian to meters
#         buildings[key] = dist*ms_per_radian
        pois_with_neigh[key] = dist*ms_per_radian
    
# heat_chosen_buildings = np.asarray(heat_chosen_buildings).T
pois_with_neigh.head(5)
```
